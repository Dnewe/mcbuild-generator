Used to create the relevant blocks list used in mcbuild_generator.processing.filter_builds.filter_irrelevant_builds

```python
import os
import numpy as np

from mcbuild_generator.utils.fs_io import read_json, write_json
from mcbuild_generator.processing.count_used_blocks import get_all_blocks


ALL_BLOCKS_JSON = "../data/09_external/all_blocks.json"
RELEVANT_BLOCK_JSON = "../data/09_external/relevant_blocks.json"
BLOCK_TO_IDX_JSON = "../data/03_processed/block_to_idx.json"
```

```python
def is_relevant(block: str) -> bool:
    return any(
        [
            ### PROCESSED BUILDING BLOCK
            # general
            "stairs" in block,
            "slab" in block,
            "door" in block,
            "trapdoor" in block,
            # wood like
            "planks" in block,
            "fence" in block,
            "sign" in block,
            "stripped" in block,
            # stone like
            "brick" in block,
            # "wall" in block,
            "polished" in block,
            "smooth" in block,
            # processed ores
            "waxed" in block,
            # others
            "wool" in block,
            "torch" in block,
            ### FUNCTIONAL BLOCK
            "bed" in block and not "bedrock" in block,
            "furnace" in block,
            "crafting_table" in block,
            "chest" in block,
            "composter" in block,
            "anvil" in block,
            "cauldron" in block,
            "enchanting_table" in block,
            "barrel" in block,
            "smoker" in block,
            ### REDSTONE BLOCK
            "button" in block,
            "pressure" in block,
            "dropper" in block,
            "redstone_lamp" in block,
            "rail" in block,
            "repeater" in block,
            "comparator" in block,
        ]
    )
```

```python
def strip_variant(block: str) -> str:
    return block.split("[")[0]
```

```python
all_blocks = get_all_blocks(dict(read_json(ALL_BLOCKS_JSON)))
all_blocks = [strip_variant(b) for b in all_blocks]
all_blocks = list(set(all_blocks))
all_blocks = np.array(all_blocks)

mask = np.array([is_relevant(b) for b in all_blocks])
relevant_blocks = all_blocks[mask]

write_json(RELEVANT_BLOCK_JSON, relevant_blocks.tolist())

print(f"All blocks count: {len(all_blocks)}")
print(f"Relevant blocks count: {len(relevant_blocks)}")
```

### Test

```python
used_blocks = list(dict(read_json(BLOCK_TO_IDX_JSON)).keys())

for used_block in used_blocks:
    if strip_variant(used_block) not in all_blocks:
        print(f"{used_block} not found in all blocks")
```
